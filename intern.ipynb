```python
from torch.utils.data import TensorDataset, DataLoader

train_dataset = TensorDataset(
    X_train_tensor,
    y_train_tensor
)

test_dataset = TensorDataset(
    X_test_tensor,
    y_test_tensor
)

train_loader = DataLoader(
    train_dataset,
    batch_size=64,
    shuffle=True
)

test_loader = DataLoader(
    test_dataset,
    batch_size=64,
    shuffle=False
)
```

```python
device = torch.device(
    "cuda" if torch.cuda.is_available() else "cpu"
)

model = BaselineMLP().to(device)

criterion = nn.CrossEntropyLoss()

optimizer = torch.optim.Adam(
    model.parameters(),
    lr=0.001
)

print(device)

#Training Loop


num_epochs = 30

for epoch in range(num_epochs):

    model.train()

    total_loss = 0

    for X_batch, y_batch in train_loader:

        X_batch = X_batch.to(device)
        y_batch = y_batch.to(device)

        optimizer.zero_grad()

        outputs = model(X_batch)

        loss = criterion(
            outputs,
            y_batch
        )

        loss.backward()

        optimizer.step()

        total_loss += loss.item()

    print(
        f"Epoch {epoch+1}/{num_epochs}, "
        f"Loss: {total_loss:.4f}"
    )
```

```python
from sklearn.metrics import (
    accuracy_score,
    f1_score,
    confusion_matrix
)

model.eval()

predictions = []
targets = []

with torch.no_grad():

    for X_batch, y_batch in test_loader:

        X_batch = X_batch.to(device)

        outputs = model(X_batch)

        preds = outputs.argmax(dim=1)

        predictions.extend(
            preds.cpu().numpy()
        )

        targets.extend(
            y_batch.numpy()
        )

acc = accuracy_score(
    targets,
    predictions
)

f1 = f1_score(
    targets,
    predictions
)

print("Accuracy:", acc)
print("F1 Score:", f1)

print(
    confusion_matrix(
        targets,
        predictions
    )
)
```

```python
import torch
import torch.nn as nn

class TextVAE(nn.Module):

    def __init__(self):

        super().__init__()

        self.encoder = nn.Sequential(
            nn.Linear(300,128),
            nn.ReLU(),
            nn.Linear(128,64),
            nn.ReLU()
        )

        self.mu = nn.Linear(64,32)
        self.logvar = nn.Linear(64,32)

        self.decoder = nn.Sequential(
            nn.Linear(32,64),
            nn.ReLU(),
            nn.Linear(64,128),
            nn.ReLU(),
            nn.Linear(128,300)
        )
  #Text VAE
    def reparameterize(self, mu, logvar):

        std = torch.exp(0.5 * logvar)

        eps = torch.randn_like(std)

        return mu + eps * std

    def forward(self,x):

        h = self.encoder(x)

        mu = self.mu(h)
        logvar = self.logvar(h)

        z = self.reparameterize(mu,logvar)

        recon = self.decoder(z)

        return recon, mu, logvar
```

```python
def vae_loss(recon_x, x, mu, logvar):

    recon_loss = nn.functional.mse_loss(
        recon_x,
        x
    )

    kl_loss = -0.5 * torch.mean(
        1 + logvar
        - mu.pow(2)
        - logvar.exp()
    )

    return recon_loss + 0.001 * kl_loss
```

```python
# Import required libraries
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader

# Convert text features to tensor
X_text_tensor = torch.tensor(
    X_text,
    dtype=torch.float32
)

# Create dataset and dataloader
text_dataset = TensorDataset(X_text_tensor)

text_loader = DataLoader(
    text_dataset,
    batch_size=64,
    shuffle=True
)


# Create VAE model
text_vae = TextVAE()

# Use GPU if available
device = torch.device(
    "cuda" if torch.cuda.is_available() else "cpu"
)

text_vae = text_vae.to(device)

# Adam optimizer
optimizer = torch.optim.Adam(
    text_vae.parameters(),
    lr=0.001
)




# Train VAE
epochs = 30

for epoch in range(epochs):

    text_vae.train()

    total_loss = 0

    for batch in text_loader:

        x = batch[0].to(device)

        optimizer.zero_grad()

        recon, mu, logvar = text_vae(x)

        loss = vae_loss(
            recon,
            x,
            mu,
            logvar
        )

        loss.backward()

        optimizer.step()

        total_loss += loss.item()

    print(
        f"Epoch {epoch+1}/{epochs} "
        f"Loss: {total_loss:.4f}"
    )
```

```python
# Switch to evaluation mode
text_vae.eval()

latent_text = []

with torch.no_grad():

    for batch in text_loader:

        x = batch[0].to(device)

        h = text_vae.encoder(x)

        mu = text_vae.mu(h)

        latent_text.append(
            mu.cpu()
        )

latent_text = torch.cat(
    latent_text,
    dim=0
)

print(latent_text.shape)

# Save latent representation
torch.save(
    latent_text,
    "latent_text.pt"
)

print("Saved Successfully")
```

```python
# Audio encoder and decoder
class AudioVAE(nn.Module):

    def __init__(self):

        super().__init__()

        self.encoder = nn.Sequential(
            nn.Linear(74, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU()
        )

        self.mu = nn.Linear(32, 16)
        self.logvar = nn.Linear(32, 16)

        self.decoder = nn.Sequential(
            nn.Linear(16, 32),
            nn.ReLU(),
            nn.Linear(32, 64),
            nn.ReLU(),
            nn.Linear(64, 74)
        )

    def reparameterize(self, mu, logvar):

        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)

        return mu + eps * std

    def forward(self, x):

        h = self.encoder(x)

        mu = self.mu(h)
        logvar = self.logvar(h)

        z = self.reparameterize(mu, logvar)

        recon = self.decoder(z)

        return recon, mu, logvar
```

```python
# Merge all modalities
latent_fusion = torch.cat(
    [
        latent_text,
        latent_audio,
        latent_visual
    ],
    dim=1
)

print(latent_fusion.shape)


# Save fused features
torch.save(
    latent_fusion,
    "latent_fusion.pt"
)

print("Fusion Saved")
```

```python
# Convert labels to tensor
y_tensor = torch.tensor(
    Y,
    dtype=torch.long
)

print(y_tensor.shape)


# Split latent features
from sklearn.model_selection import train_test_split

indices = np.arange(len(Y))

train_idx, test_idx = train_test_split(
    indices,
    test_size=0.2,
    random_state=42,
    stratify=Y
)

X_train_latent = latent_fusion[train_idx]
X_test_latent = latent_fusion[test_idx]

y_train = y_tensor[train_idx]
y_test = y_tensor[test_idx]

print(X_train_latent.shape)
print(X_test_latent.shape)


# Create dataloaders
train_dataset = TensorDataset(
    X_train_latent,
    y_train
)

test_dataset = TensorDataset(
    X_test_latent,
    y_test
)

train_loader = DataLoader(
    train_dataset,
    batch_size=64,
    shuffle=True
)

test_loader = DataLoader(
    test_dataset,
    batch_size=64,
    shuffle=False
)




# Classifier on latent space
class FusionClassifier(nn.Module):

    def __init__(self):

        super().__init__()

        self.net = nn.Sequential(

            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Dropout(0.3),

            nn.Linear(32, 16),
            nn.ReLU(),

            nn.Linear(16, 2)
        )

    def forward(self, x):

        return self.net(x)



# Initialize model
fusion_model = FusionClassifier().to(device)

criterion = nn.CrossEntropyLoss()

optimizer = torch.optim.Adam(
    fusion_model.parameters(),
    lr=0.001
)

epochs = 30

for epoch in range(epochs):

    fusion_model.train()

    total_loss = 0

    for X_batch, y_batch in train_loader:

        X_batch = X_batch.to(device)
        y_batch = y_batch.to(device)

        optimizer.zero_grad()

        outputs = fusion_model(X_batch)

        loss = criterion(
            outputs,
            y_batch
        )

        loss.backward()

        optimizer.step()

        total_loss += loss.item()

    print(
        f"Epoch {epoch+1}/{epochs} "
        f"Loss: {total_loss:.4f}"
    )
```

```python
from sklearn.metrics import (
    accuracy_score,
    f1_score,
    confusion_matrix
)

fusion_model.eval()

predictions = []
targets = []

with torch.no_grad():

    for X_batch, y_batch in test_loader:

        X_batch = X_batch.to(device)

        outputs = fusion_model(X_batch)

        preds = outputs.argmax(dim=1)

        predictions.extend(
            preds.cpu().numpy()
        )

        targets.extend(
            y_batch.numpy()
        )

acc = accuracy_score(
    targets,
    predictions
)

f1 = f1_score(
    targets,
    predictions
)

print("Accuracy:", acc)
print("F1 Score:", f1)

print(
    confusion_matrix(
        targets,
        predictions
    )
)
```

```python
# Reconstruct audio latent
class AudioReconstructor(nn.Module):

    def __init__(self):

        super().__init__()

        self.net = nn.Sequential(

            nn.Linear(48, 64),
            nn.ReLU(),

            nn.Linear(64, 32),
            nn.ReLU(),

            nn.Linear(32, 16)
        )

    def forward(self, x):

        return self.net(x)
```

```python
# Input = text + visual
X_reconstruct = torch.cat(
    [
        latent_text,
        latent_visual
    ],
    dim=1
)

# Target = audio
y_reconstruct = latent_audio

print(X_reconstruct.shape)
print(y_reconstruct.shape)

recon_dataset = TensorDataset(
    X_reconstruct,
    y_reconstruct
)

recon_loader = DataLoader(
    recon_dataset,
    batch_size=64,
    shuffle=True
)


audio_reconstructor = AudioReconstructor().to(device)

criterion = nn.MSELoss()

optimizer = torch.optim.Adam(
    audio_reconstructor.parameters(),
    lr=0.001
)

epochs = 30

for epoch in range(epochs):

    total_loss = 0

    for x_batch, y_batch in recon_loader:

        x_batch = x_batch.to(device)
        y_batch = y_batch.to(device)

        optimizer.zero_grad()

        pred = audio_reconstructor(
            x_batch
        )

        loss = criterion(
            pred,
            y_batch
        )

        loss.backward()

        optimizer.step()

        total_loss += loss.item()

    print(
        f"Epoch {epoch+1} "
        f"Loss: {total_loss:.4f}"
    )
```

```python
# Generate audio latent
reconstructed_audio = audio_reconstructor(
    torch.cat(
        [
            latent_text,
            latent_visual
        ],
        dim=1
    ).to(device)
)

reconstructed_audio = reconstructed_audio.cpu()


# Replace missing audio
latent_recovered = torch.cat(
    [
        latent_text,
        reconstructed_audio,
        latent_visual
    ],
    dim=1
)

print(latent_recovered.shape)



outputs = fusion_model(
    latent_recovered[test_idx].to(device)
)

preds = outputs.argmax(
    dim=1
).cpu().numpy()

diffusion_acc = accuracy_score(
    y_test.numpy(),
    preds
)

diffusion_f1 = f1_score(
    y_test.numpy(),
    preds
)

print("Accuracy:", diffusion_acc)
print("F1:", diffusion_f1)


results_df.loc[
    results_df["Model"] == "Diffusion Reconstruction",
    ["Accuracy", "F1"]
] = [diffusion_acc, diffusion_f1]

print(results_df)
```
